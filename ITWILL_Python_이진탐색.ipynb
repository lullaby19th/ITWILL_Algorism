## ■ 이진탐색 알고리즘(week2)

### ※ 이진탐색이란

정렬된 데이터를 좌우 둘로 나눠서 원하는 값의 탐색범위를 좁혀가며 찾는 방법

## ■ 문제. 다음 리스트의 개수가 홀수일 때 이진탐색을 구현해보시오. 

a = [1, 7, 11 ,12, 14, 23, 33, 47, 51, 64, 67, 77, 139, 672, 871]

```python
#%% Step1: 알고리즘 day2 이진검색 list개수 홀수일때
a = [1,7,11,12,14,23,33,47,51,64,67,77,139,672,871]
b = int(input('검색할 숫자 입력하세요~'))
print(type(a))


for i in range(1,len(a)): 
# range범위는 큰 의미없음 그냥넉넉하게 반복문 돌리려고하는거임.
# 따라서 while true: 로 무한루프 돌려도 됨.
    
    half = int(len(a)/2) #리스트의 중앙값의 위치 len(a)/2에다가 int로 정수형만들기
    
    if b > a[half]: # 찾고자 하는 값이 중앙값보다 클 때
        del (a[0:half+1]) #처음부터 중간까지 cut
    
    elif b < a[half]: # 찾고자 하는 값이 중앙값보다 클 때
        del (a[half:]) # 중간부터 끝까지 cut
    
    else: # b = a[half] 찾고자 하는 값이 중앙값이랑 같을 떄
        del(a[0:half]) #왼쪽 사이드 다 cut
        del(a[1:]) #오른쪽 사이드 다 cut
        break

print(a)
```
