```python
# test case는 다 통과하나, 프로그래머스로는 통과 못하는 코드임( 뜯어 고쳐야됨. )

def solution(n, t, m, timetable):
    answer = ''
    start = 540
    num = 0
    tm = []
    for i in timetable:
        minute = (60*int(i[0:2])) + int(i[3:])
        tm.append(minute)
    print(tm)
    tm_s = sorted(tm)
    
    if len(timetable) < m:
        num = start + t*(n-1)
        
    elif n * m == len(timetable):
        if tm_s[-1] > start + t*(n-1):
            num = start + t*(n-1)
        else:
            num = tm_s[-1] -1
            
    elif n * m > len(timetable):
        if tm_s[-1] > start + t*(n-1):
            num = start + t*(n-1)
        else:
            num = tm_s[m] - 1
    
    num1 = num // 60
    num2 = num % 60
    x = str(num1).rjust(2,'0')
    y = str(num2).rjust(2,'0')
    
    print(str(num1).rjust(2,'0'))
    print(str(num2).rjust(2,'0'))
    
    tmp1 = [x, y]
    bond = ':'
    
    result = bond.join(tmp1)
    answer += result
    return answer
```
