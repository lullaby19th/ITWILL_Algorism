```python
# Def를 써서 마무리 .ver2

import math
from collections import defaultdict

def knn(t,k):
    x = [8,2,7,7,3,1]
    y = [5,3,10,3,8,1]
    label = ['과일', '단백질', '채소', '과일', '채소', '단백질']
    label_s = list(set(label)) # 중복된거 제거
    
    d = []
    box = defaultdict(list)
    
    for i in range(len(x)):
        x_t = (t[0] - x[i]) ** 2
        y_t = (t[1] - y[i]) ** 2
        result = round(math.sqrt(x_t + y_t),k)
        d.append(result) # 유클리드 기하학으로 나온거리를 리스트로 다 나타냄.
    
    s_d = sorted(d)[0:k] # k까지 순위 정해서 추출.
    
    for i, label in enumerate(label):
        box[label].append(d[i])
    
    li = []
    
    for i in box.keys():
        for j in range (len(box[i])):    
            if box[i][j] in s_d:
                li.append(i)
    
    cnt = [li.count(i) for i in label_s ]
    answer = [i for i,z in zip(label_s,cnt) if z == max(cnt)]

    return answer[0]

tomato = [6,4]

print ( knn(tomato,k=3) )
```

### ※ 다른 모범 답안

```python
# counter 활용 
# Dictionary 자료형도 sort적용가능 -> sort 적용시 key값 기준으로 정렬이 된다.

from collections import Counter

x = [8,2,7,7,3,1]

y = [5,3,10,3,8,1]

label=['과일','단백질','채소','과일','채소','단백질']

tomato =[6,4]



def knn(f,k) :

    dist = [(((f[0]-x[i])**2) + ((f[1]-y[i])**2))**0.5 for i in range(len(x))] #거리 계산

    box = dict(zip(dist, label))

    box = sorted(box.items()) #dict만든 후 정렬

    rank_k = [j[1] for j in box[:k]] #k개수만큼 추출
    cnt = Counter(rank_k) #요소별 개수(자동 내림차순)
    
    return list(cnt.keys())[0] #개수 제일 많은 것 출력

print ( knn(tomato,k=3) )
```

```python
# list형의 count 함수와 index 함수를 활용

def euclid_dist(x,y,z,k):

    z_x=z[0]

    z_y=z[1]

    dist=[]

    for i in range(len(x)):

        res=round(((x[i]-z_x)**2 + (y[i]-z_y)**2)**0.5,2)

        dist.append(res)

    nearest=[]

    dist_sorted=sorted(dist)

    for j in range(k):

        nearest.append(dist.index(dist_sorted[j]))        

    return nearest


def knn(z,k):

    category=[]

    for i in euclid_dist(x,y,z,k):

        category.append(label[i])    

    fruit=category.count('과일')

    protein=category.count('단백질')

    veg=category.count('채소')



    if fruit>protein: return '과일'

    elif fruit<protein: return '단백질'

    elif veg>fruit: return '채소'

    elif veg<fruit: return '과일'

    elif veg>protein: return '채소'

    elif veg<protein: return '단백질'



x = [ 8, 2, 7, 7, 3, 1 ]

y = [ 5, 3, 10, 3, 8, 1 ]

label = [ '과일', '단백질', '채소', '과일', '채소', '단백질' ]

tomato=[ 6,4 ]

print(knn(tomato,3))
```
