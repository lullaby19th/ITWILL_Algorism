### 2017년 카카오 블라인드 코딩시험 5번째 문제

## ■ 문제. 다음 예제 입출력에 대하여 다음 결과가 나오도록 코딩하시오!

|    str1   |     str2    | answer |
|:---------:|:-----------:|:------:|
|   FRANCE  |    french   |  16384 |
| handshake | shake hands |  65536 |
|  aa1+aa2  |    AAAA12   |  43690 |
|  E+M*C^2  |   e=m*c^2   |  65536 |

```python
# 일반적인 경우 test.ver

import re # re.sub를 사용하기 위해 불러오는 모듈

str1 = "aa1+aa2"
str2 = "AAAA12"
a = []
b = []

# 두글자씩 리스트 요소로 가지기 위해서 쓰는 코드
for i in range(len(str1)-1):
    a.append(str1[i:i+2])  

for i in range(len(str2)-1):
    b.append(str2[i:i+2])

# 특수문자, 숫자, 공백 제외하고 리스트 요소로 추가하기 위해 쓰는 코드
c = []
d = []
for i in a:
    c.append(re.sub('[^"a-z""A-Z"]', '', i)) # ^ = not
    

for i in b:
    d.append(re.sub('[^"a-z""A-Z"]', '', i)) # ^ = not

# 두글자로 구성된 요소가 아닌것들은 제외함. (len(c) ==2 , len(d) == 2 )
x = []
y = []

for i in range(len(c)):
    if len(c[i]) == 2:
        x.append(c[i])

for i in range(len(d)):
    if len(d[i]) == 2:
        y.append(d[i])

cnt = 0 

print(x)
print(y)


for i in range(len(x)):
    for j in range(len(y)):
        if y[j].lower() == x[i].lower():
            cnt +=1
            break # 교집합을 한번 더 세는것을 방지하기 위해서 break
            
if cnt == 0: # 교집합이 존재하지 않을 시 (공집합)
    result = 1
    
else : # 교집합이 존재할 시
    result = cnt/ (len(x)+len(y)-cnt)


answer = int(result * 65536) # 정수로 표현해야되므로 int

print(answer)
```

```python
#%% def로 표현 final ver.

def jacard(str1, str2):
    import re
    a = []
    b = []
    
    for i in range(len(str1)-1):
        a.append(str1[i:i+2])
    
    for i in range(len(str2)-1):
        b.append(str2[i:i+2])
    
    c = []
    d = []
    for i in a:
        c.append(re.sub('[^"a-z""A-Z"]', '', i))
            
    for i in b:
        d.append(re.sub('[^"a-z""A-Z"]', '', i))
    
    x = []
    y = []
    
    for i in range(len(c)):
        if len(c[i]) == 2:
            x.append(c[i])
    
    for i in range(len(d)):
        if len(d[i]) == 2:
            y.append(d[i])
    
    cnt = 0 
    for i in range(len(x)):
        for j in range(len(y)):
            if y[j].lower() == x[i].lower():
                cnt +=1
                break
    if cnt == 0:
        result = 1
        
    else :
        result = cnt/ (len(x)+len(y)-cnt)
    
    answer = int(result * 65536)
    return answer

print(jacard('FRANCE','french'))

print(jacard('handshake','shake hands'))

print(jacard('aa1+aa2','AAAA12'))

print(jacard('E=M*C^2','e=m*c^2'))
```
