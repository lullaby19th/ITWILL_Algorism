```python
# def까지 써서 완성 .ver2

import pandas as pd 
import numpy as np
df = pd.read_csv('D:\\data\\skin.csv')

def tree():
    x = list(df)[1:-1] # gender에서 car까지
    y = list(df)[-1] # cupon_react만 출력
    before = []
    after = []
    
    # 분할전과 분할후 값을 구하기전에 crosstab을 수행함.
    for i in x:
        xl = np.unique(df[i].values) # 'NO', 'YES'
        yl = np.unique(df[y].values) # 'NO', 'YES'
        ct = pd.crosstab(df[y], df[i], margins = True)
        print(ct)
        t = ct['All']['All']
        
        # 분할전 
        r = 0
        for j in xl:
            h = 0.001
            r += (ct[j]['All'] / t) * np.log2((ct[j]['All'] / t)+h)
            result1 = -r
        before.append(result1)
        
        # 분할후
        tmp = 0
        for i in yl:
            tt = ct['All'][i] / t
            for j in xl:
                h = 0.001
                tmp += tt * ( (ct[j][i] / ct['All'][i]) * np.log2((ct[j][i] / ct['All'][i])+h) )
        after.append(-tmp)
    
    # 정보획득량을 구해서 리스트에 담음
    answer = [round(i-j,3) for i,j in zip(before,after)]
    
    for i,j in zip(x,answer):
        print ('{0} 의 정보획득량 은 = {1}'.format(i,j))
        
    return '정보 획득량 산출 끝'

print(tree())
```
